This project applies ensemble machine learning models to predict credit card
default risk. The workflow covers data understanding, ensemble modeling,
feature importance, and error analysis with a focus on model reliability.


### Models Evaluated

- Logistic Regression (baseline reference)
- Random Forest
- Gradient Boosting

Models were evaluated using ROC AUC, recall, precision, and confusion matrices
to reflect the cost-sensitive nature of credit risk.


### Performance Comparison

Among all evaluated models, Gradient Boosting demonstrated the strongest
overall performance, particularly in identifying high-risk default cases.
Random Forest provided stable and competitive results, while Logistic
Regression served as a useful baseline but lacked the capacity to model
non-linear relationships.


### Feature Importance Summary

Payment history features were the strongest predictors of default risk.
Credit limit and bill amounts provided secondary signals, while demographic
features played a smaller role.


### Error Analysis Insights

Although Gradient Boosting improved recall for defaulters, some false negatives
persisted near the decision threshold. Most errors occurred in borderline
cases, indicating uncertainty rather than systematic model failure.


### Model Selection Decision

Gradient Boosting was selected as the final model due to its superior ability
to capture complex patterns in customer repayment behavior. By focusing
sequentially on difficult-to-classify cases, the model achieved stronger
recall for defaulters, which is critical in credit risk prediction.

While Gradient Boosting requires careful control to avoid overfitting, its
performance advantages justify its selection in risk-sensitive decision
contexts.


### Limitations & Future Work

- Class imbalance affects recall stability
- Feature importance is not causal
- Future work includes threshold optimization, calibration, and ensemble stacking


### Final Takeaways

- Gradient Boosting effectively captures complex credit risk patterns.
- Boosting improves recall for defaulters by focusing on difficult cases.
- Model complexity requires careful tuning to prevent overfitting.
- Business-aligned evaluation is essential in credit risk modeling.
